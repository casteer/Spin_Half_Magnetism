```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3,3,100)
y = np.tanh(x)

plt.figure(figsize=(8, 6), dpi=100, facecolor='w', edgecolor='w')
plt.plot(x,y,'r-',linewidth=4)
plt.grid()
plt.xlim([0,3])
plt.ylim([0,1])
plt.xlabel(r'$\frac{\mu_{B} H}{k_{B}T}$', fontsize=18)
plt.ylabel(r'$\frac{\langle \mu \rangle}{\mu_{B}}$', fontsize=18)

plt.text(1.4,0.2,r'$M = N\langle \mu \rangle  = N \, \mu_{B} \, \tanh \left( \frac{\mu_{B} H}{k_{B}T} \right)$',fontsize=18, bbox=dict(facecolor='w',edgecolor='w'))
```

```python
# Solving the spin 1/2 magnetization for Weiss' model 

%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

from cycler import cycler

from scipy.optimize import fsolve

x = np.linspace(0,3,100)


plt.figure(figsize=(8, 6), dpi=100, facecolor='w', edgecolor='w')
plt.plot(x,x,'k:',linewidth=2, label = 'M, left-hand side')

a_scale = [0.5,1,1.1,1.2,1.4,1.6,1.8,2]
c = cycler('color', plt.cm.viridis(np.linspace(0, 1, len(a_scale))) )
plt.rc('axes', prop_cycle=(c))    

for a in a_scale:
    
    initial_guess = 1
    def dM(x):
        return x - np.tanh(a*x)
    xs = fsolve(dM, initial_guess)
    
    ys = np.tanh(a*xs)
#     print(xs,ys)
    y = np.tanh(a*x)

    graph_label = "a = %d" % (a)
    plt.plot(x,y,'-',linewidth=1)
    plt.plot(xs,ys,'ko')
    

plt.plot([0],[0],'-',linewidth=1,label='M, right-hand side')
plt.plot(xs,ys,'ko',label='Magnetization Solution')
    

plt.grid()
# plt.xlim([0,3])
plt.ylim([0,1])
plt.xlabel(r'$\frac{\mu_{B} M }{k_{B}T}$', fontsize=18)
plt.ylabel(r'$\frac{M}{n \mu_{B}}$', fontsize=18)
plt.legend()

plt.text(1.8,0.3,r'$M = n \, \mu_{B} \, \tanh \left( \frac{\mu_{B} \alpha }{k_{B}T} M \right)$',fontsize=18, bbox=dict(facecolor='w',edgecolor='w'))
plt.text(0.0,0.9,r'Increasing $\frac{ \alpha }{ T}  $',fontsize=12, bbox=dict(facecolor='w',edgecolor='w'))

ax = plt.axes()
ax.arrow(0.4, 0.8, -0.1, 0.05,head_width=0.02, head_length=0.02,length_includes_head=True)

```
