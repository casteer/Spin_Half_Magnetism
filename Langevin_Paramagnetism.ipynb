```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

from cycler import cycler

x = np.linspace(0,10,1000)
y = (1.0/np.tanh(x)) - (1.0/x)

plt.figure(figsize=(8, 6), dpi=100, facecolor='w', edgecolor='w')
plt.plot(x,y,'r-',linewidth=4, label=str('Classical (Langevin)'))
# plt.xlim([0,3])
plt.grid()
plt.ylim([0,1])
plt.xlabel(r'$\frac{\mu_{B} H}{k_{B}T}$', fontsize=18)
plt.ylabel(r'$\frac{\langle \mu \rangle}{\mu}$', fontsize=18)

plt.figure(figsize=(8, 6), dpi=100, facecolor='w', edgecolor='w')
plt.plot(x,y,'k:',linewidth=4, label=str('Classical (Langevin)'))
plt.rc('axes', prop_cycle=(cycler('color', ['r', 'g', 'b', 'y'])))
for J in  [0.5, 1.5, 3.5]: 
    J0 = (2*J+1)/(2*J)
    J1 = (1/(2*J))
    B = (J0/np.tanh(J0*x)) - (J1/np.tanh(J1*x))
    graph_label = "J = %3.1f" % (J)
    plt.plot(x, B,'-',linewidth=4, label=graph_label)

plt.grid()
plt.legend()
# plt.xlim([0,3])
# plt.ylim([0,1])
plt.xlabel(r'$\frac{\mu_{B} H}{k_{B}T}$', fontsize=18)
plt.ylabel(r'$\frac{\langle \mu \rangle}{g_{J} \; \; J \; \; \mu_{B}}$', fontsize=18)




# plt.text(14,0.2,r'$M = N\langle \mu \rangle  = N \, \mu_{B} \, \tanh \left( \frac{\mu_{B} H}{k_{B}T} \right)$',fontsize=18, bbox=dict(facecolor='w',edgecolor='w'))
```
